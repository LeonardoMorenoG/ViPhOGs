```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Import time tables
tFamily = pd.read_csv("timesFamily.wall.csv",sep=' ')
tOrder = pd.read_csv("timesOrder.wall.csv",sep=' ')
tFamily2 = pd.read_csv("Family/timesFamily.wall.csv",sep=' ')
tGenus = pd.read_csv("Family/timesGenus.wall.csv",sep=' ')
```

```python
def getDispersionValues(table):
    return (table.groupby("n_estimators").mean(),table.groupby("n_estimators").std())
```
