```python
import MySQLdb
import plotly
import plotly.graph_objs as go
plotly.offline.init_notebook_mode() # run at the start of every notebook
```

# 1. Genomes length sorted by accession

```python
cursor = db.cursor()
query = ("SELECT accession, length "
         "FROM bioentry INNER JOIN biosequence "
         "WHERE bioentry.bioentry_id=biosequence.bioentry_id AND bioentry.biodatabase_id=1 "
#         "ORDER BY accession")
         "ORDER BY length")
cursor.execute(query)
```

```python
lengths = []
accessions = []
for accession,length in cursor:
    accessions.append(accession)
    lengths.append(length)
cursor.close()
print lengths[0:10]
```

```python
trace1 = go.Bar(
    x=accessions,
    y=lengths
)
data = [trace1]
layout = go.Layout(title='Genomes Length')
fig = go.Figure(data=data, layout=layout)
plotly.offline.iplot(fig)
```

# 2. Number of proteins sorted by accession

```python
cursor = db.cursor()
query = ("SELECT accession, COUNT(seqfeature_id) as l "
         "FROM bioentry INNER JOIN seqfeature  "
         "WHERE seqfeature.bioentry_id=bioentry.bioentry_id AND seqfeature.type_term_id!=10 "
         "AND bioentry.biodatabase_id=2 "
         "GROUP BY accession ORDER BY accession ")
cursor.execute(query)

accessions = []
numProteins = []
for accession, proteins in cursor:
    accessions.append(accession)
    numProteins.append(proteins)
cursor.close()
print accessions[0:10]
print numProteins[0:10]
```

```python
cursor = db.cursor()
query = ("SELECT bioentry.accession, COUNT(seqfeature.seqfeature_id)"
         "FROM bioentry INNER JOIN biosequence INNER JOIN seqfeature "
         "WHERE bioentry.bioentry_id=biosequence.bioentry_id AND bioentry.bioentry_id=seqfeature.bioentry_id "
         "AND bioentry.biodatabase_id=3 "
         "GROUP BY bioentry.accession ")
cursor.execute(query)

accessionsV = []
numProteinsV = []
for accession, proteins in cursor:
    accessionsV.append(accession)
    numProteinsV.append(proteins)
cursor.close()
print accessionsV[0:10]
print numProteinsV[0:10]
```

```python
accessionsFinal = accessions + accessionsV
numProteinsFinal = numProteins + numProteinsV
trace1 = go.Bar(
    x=accessionsFinal,
    y=numProteinsFinal
)
data = [trace1]
layout = go.Layout(title='Proteins per Genome')
fig = go.Figure(data=data, layout=layout)
plotly.offline.iplot(fig)
```
